# Tokenization:-

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
sen_token=nltk.sent_tokenize(content)
sen_token
```

```python
w_sent=word_tokenize(content)
w_sent
```

```python
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
ps=PorterStemmer()
wnt=WordNetLemmatizer()
```

```python
import re
```

```python
corpus=[]

for i in range(len(sen_token)):
    punct=re.sub("[^A-Za-z0-9]"," ",sen_token[i])
    punct_low=punct.lower()
    punct_split=punct_low.split()
    punct_stop=[wnt.lemmatize(word) for word in punct_split if not word in (stopwords.words('english'))]
    punct_join=' '.join(punct_stop)
    corpus.append(punct_join)
print(corpus)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
cv=CountVectorizer()
bow=cv.fit_transform(corpus).toarray()
bow
```

```python
tf=TfidfVectorizer()
tfid=tf.fit_transform(corpus).toarray()
tfid
```
